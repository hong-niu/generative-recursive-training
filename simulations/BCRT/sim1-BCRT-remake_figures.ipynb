### Re-compute W1 and MMD rates from saved run

```python
#!/usr/bin/env python3

from pathlib import Path
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


# ============================================================
# User settings
# ============================================================

RUN_DIR = Path("logs/KDE/biased_alphaReal_h0=2_T=3000_m1=50_nreps=100_20260611-214856")

NEW_DROP_FIRST = 200 # same as experiment 
OUT_SUMMARY_CSV = RUN_DIR / f"summary_table_recomputed_drop{NEW_DROP_FIRST}.csv"
OUT_REP_CSV = RUN_DIR / f"all_rep_index_recomputed_drop{NEW_DROP_FIRST}.csv"
OUT_RESULTS_JSON = RUN_DIR / f"all_results_recomputed_drop{NEW_DROP_FIRST}.json"

# New output folder for only the last repetition per combo
OUT_LAST_REP_RATE_DIR = RUN_DIR / f"rate_plots_last_reps_recomputed_drop{NEW_DROP_FIRST}"


# ============================================================
# Slope fitting
# ============================================================

def estimate_power_law_slope(M, vals, drop_first=300):
    M = np.asarray(M, dtype=float)
    vals = np.asarray(vals, dtype=float)

    if drop_first >= len(M) - 5:
        drop_first = max(0, len(M) - 5)

    M_use = M[drop_first:]
    vals_use = vals[drop_first:]

    mask = (
        np.isfinite(M_use)
        & np.isfinite(vals_use)
        & (M_use > 0)
        & (vals_use > 0)
    )

    M_use = M_use[mask]
    vals_use = vals_use[mask]

    if M_use.size < 3:
        raise ValueError(
            f"Not enough valid positive points to fit slope after drop_first={drop_first}."
        )

    logM = np.log(M_use).reshape(-1, 1)
    logV = np.log(vals_use)

    reg = LinearRegression().fit(logM, logV)
    return float(reg.coef_[0]), float(reg.intercept_)


def resolve_saved_path(path_str):
    """
    Your original all_rep_index.csv may contain paths relative to the directory
    where the experiment was launched. This tries the path directly first,
    then tries relative to RUN_DIR.
    """
    path_str = str(path_str).strip()

    if not path_str or path_str.lower() == "nan":
        return None

    p = Path(path_str)

    if p.exists():
        return p

    p2 = RUN_DIR / p
    if p2.exists():
        return p2

    return None


def read_rep_metrics(rep_row):
    """
    Prefer metrics_csv_path because it is lighter.
    Fall back to npz_path if needed.
    """
    metrics_path = resolve_saved_path(rep_row.get("metrics_csv_path", ""))
    npz_path = resolve_saved_path(rep_row.get("npz_path", ""))

    if metrics_path is not None:
        df = pd.read_csv(metrics_path)

        M = df["M"].to_numpy(float)

        # These are the exact arrays your original code used for slope fitting.
        # They already include the /log(M) correction when p == alpha_real.
        W1_for_slope = df["W1_for_slope"].to_numpy(float)
        MMD_for_slope = df["MMD_for_slope"].to_numpy(float)

        return M, W1_for_slope, MMD_for_slope

    if npz_path is not None:
        z = np.load(npz_path, allow_pickle=True)

        M = z["M"].astype(float)
        W1_for_slope = z["W1_for_slope"].astype(float)
        MMD_for_slope = z["MMD_for_slope"].astype(float)

        return M, W1_for_slope, MMD_for_slope

    raise FileNotFoundError(
        f"Could not find metrics_csv_path or npz_path for rep row:\n{rep_row}"
    )


# ============================================================
# Plotting
# ============================================================

def plot_with_trend(
    M,
    values,
    metric_name,
    p_val,
    alpha_real_val,
    bias_rate,
    rep_idx=None,
    drop_first=300,
    save_path=None,
):
    M = np.asarray(M, dtype=float)
    values = np.asarray(values, dtype=float)

    valid = np.isfinite(M) & np.isfinite(values) & (M > 0) & (values > 0)

    log_M = np.full_like(M, np.nan, dtype=float)
    log_val = np.full_like(values, np.nan, dtype=float)

    log_M[valid] = np.log10(M[valid])
    log_val[valid] = np.log10(values[valid])

    plt.figure(figsize=(6, 4))
    plt.plot(log_M, log_val, "o-", alpha=0.5, markersize=3, label="Distr. Loss")

    fit_mask = valid.copy()
    fit_mask[:drop_first] = False

    if np.count_nonzero(fit_mask) > 5:
        x_fit = log_M[fit_mask]
        y_fit = log_val[fit_mask]

        slope, intercept = np.polyfit(x_fit, y_fit, 1)
        y_pred = slope * x_fit + intercept

        plt.plot(
            x_fit,
            y_pred,
            "r--",
            linewidth=2,
            label=f"Fit slope: {slope:.3f}; rate: {-slope:.3f}",
        )

    plt.xlabel(r"$\log_{10}$(M$_n$)")

    if metric_name == "W1":
        metric_name_tex = r"W_1"
    elif metric_name == "MMD":
        metric_name_tex = r"\mathrm{MMD}"
    else:
        metric_name_tex = metric_name

    plt.ylabel(rf"$\log_{{10}}({metric_name_tex})$")

    title = (
        rf"${metric_name_tex}$ Scaling: "
        rf"p={p_val:.2f}, $\alpha_{{real}}$={alpha_real_val:.2f}, q={bias_rate:.2f}"
    )

    if rep_idx is not None:
        title += rf", rep={int(rep_idx):03d}"

    title += rf", drop={drop_first}"

    plt.title(title)
    plt.legend()
    plt.grid(True, which="both", ls="--", alpha=0.3)
    plt.tight_layout()

    if save_path is not None:
        save_path = Path(save_path)
        save_path.parent.mkdir(parents=True, exist_ok=True)
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
        plt.close()
    else:
        plt.show()


def combo_dirname(p_val, alpha_real_val, bias_rate):
    return (
        f"p={p_val:.2f}_alphaReal={alpha_real_val:.2f}_q={bias_rate:.2f}"
        .replace("=", "")
        .replace(".", "p")
    )


def remake_last_rep_rate_plots(recomputed_rep_df):
    """
    For each alpha/q/p combo, find the last rep, remake W1 and MMD rate plots,
    and save them in OUT_LAST_REP_RATE_DIR.
    """
    OUT_LAST_REP_RATE_DIR.mkdir(parents=True, exist_ok=True)

    group_cols = [
        "alpha_real",
        "alpha_est",
        "bias_rate",
        "p_target",
        "T",
        "m1",
        "m2",
        "theoretical_rate",
    ]

    saved_plot_rows = []

    for keys, g in recomputed_rep_df.groupby(group_cols, dropna=False):
        key_dict = dict(zip(group_cols, keys))

        # Last rep means highest rep index within this combo
        last_row = g.sort_values("rep").iloc[-1].to_dict()
        rep_idx = int(last_row["rep"])

        M, W1_for_slope, MMD_for_slope = read_rep_metrics(last_row)

        p_val = float(last_row["p_target"])
        alpha_real_val = float(last_row["alpha_real"])
        bias_rate = float(last_row["bias_rate"])

        dirname = combo_dirname(p_val, alpha_real_val, bias_rate)
        combo_plot_dir = OUT_LAST_REP_RATE_DIR / dirname
        combo_plot_dir.mkdir(parents=True, exist_ok=True)

        w1_plot_path = combo_plot_dir / (
            f"W1_lastrep={rep_idx:03d}_p={p_val:.2f}_alphaReal={alpha_real_val:.2f}_q={bias_rate:.2f}_drop{NEW_DROP_FIRST}.png"
        )
        mmd_plot_path = combo_plot_dir / (
            f"MMD_lastrep={rep_idx:03d}_p={p_val:.2f}_alphaReal={alpha_real_val:.2f}_q={bias_rate:.2f}_drop{NEW_DROP_FIRST}.png"
        )

        plot_with_trend(
            M=M,
            values=W1_for_slope,
            metric_name="W1",
            p_val=p_val,
            alpha_real_val=alpha_real_val,
            bias_rate=bias_rate,
            rep_idx=rep_idx,
            drop_first=NEW_DROP_FIRST,
            save_path=w1_plot_path,
        )

        plot_with_trend(
            M=M,
            values=MMD_for_slope,
            metric_name="MMD",
            p_val=p_val,
            alpha_real_val=alpha_real_val,
            bias_rate=bias_rate,
            rep_idx=rep_idx,
            drop_first=NEW_DROP_FIRST,
            save_path=mmd_plot_path,
        )

        saved_plot_rows.append({
            **key_dict,
            "rep": rep_idx,
            "W1_last_rep_rate_plot_path": str(w1_plot_path),
            "MMD_last_rep_rate_plot_path": str(mmd_plot_path),
        })

    saved_plots_df = pd.DataFrame(saved_plot_rows)
    saved_plots_csv = OUT_LAST_REP_RATE_DIR / "last_rep_rate_plot_index.csv"
    saved_plots_df.to_csv(saved_plots_csv, index=False)

    print(f"\nSaved last-rep rate plots to:\n  {OUT_LAST_REP_RATE_DIR}")
    print(f"Saved last-rep plot index to:\n  {saved_plots_csv}")


# ============================================================
# Main recomputation
# ============================================================

def main():
    rep_index_path = RUN_DIR / "all_rep_index.csv"
    if not rep_index_path.exists():
        raise FileNotFoundError(f"Could not find {rep_index_path}")

    rep_df = pd.read_csv(rep_index_path)

    recomputed_rep_rows = []

    for _, row in rep_df.iterrows():
        row_dict = row.to_dict()

        M, W1_for_slope, MMD_for_slope = read_rep_metrics(row_dict)

        slope_W1, intercept_W1 = estimate_power_law_slope(
            M, W1_for_slope, drop_first=NEW_DROP_FIRST
        )
        slope_MMD, intercept_MMD = estimate_power_law_slope(
            M, MMD_for_slope, drop_first=NEW_DROP_FIRST
        )

        new_row = dict(row_dict)
        new_row["drop_first_for_fit"] = NEW_DROP_FIRST
        new_row["slope_W1"] = slope_W1
        new_row["slope_MMD"] = slope_MMD
        new_row["intercept_W1"] = intercept_W1
        new_row["intercept_MMD"] = intercept_MMD
        new_row["observed_rate_W1"] = -slope_W1
        new_row["observed_rate_MMD"] = -slope_MMD

        recomputed_rep_rows.append(new_row)

    recomputed_rep_df = pd.DataFrame(recomputed_rep_rows)
    recomputed_rep_df.to_csv(OUT_REP_CSV, index=False)

    group_cols = [
        "alpha_real",
        "alpha_est",
        "bias_rate",
        "p_target",
        "T",
        "m1",
        "m2",
        "theoretical_rate",
    ]

    summary_rows = []

    for keys, g in recomputed_rep_df.groupby(group_cols, dropna=False):
        key_dict = dict(zip(group_cols, keys))

        rates_W1 = g["observed_rate_W1"].to_numpy(float)
        rates_MMD = g["observed_rate_MMD"].to_numpy(float)

        n_reps = len(g)

        summary_rows.append({
            **key_dict,
            "n_reps": int(n_reps),
            "drop_first_for_fit": NEW_DROP_FIRST,

            "observed_rate_mean_W1": float(np.nanmean(rates_W1)),
            "observed_rate_median_W1": float(np.nanmedian(rates_W1)),
            "observed_rate_std_W1": float(np.nanstd(rates_W1, ddof=1)) if n_reps > 1 else 0.0,

            "observed_rate_mean_MMD": float(np.nanmean(rates_MMD)),
            "observed_rate_median_MMD": float(np.nanmedian(rates_MMD)),
            "observed_rate_std_MMD": float(np.nanstd(rates_MMD, ddof=1)) if n_reps > 1 else 0.0,
        })

    summary_df = pd.DataFrame(summary_rows)

    summary_df = summary_df.sort_values(
        ["alpha_real", "bias_rate", "p_target"]
    ).reset_index(drop=True)

    summary_df.to_csv(OUT_SUMMARY_CSV, index=False)

    all_results = []

    for _, s in summary_df.iterrows():
        mask = np.ones(len(recomputed_rep_df), dtype=bool)

        for c in group_cols:
            mask &= recomputed_rep_df[c].to_numpy() == s[c]

        g = recomputed_rep_df.loc[mask]

        all_results.append({
            "alpha_real": float(s["alpha_real"]),
            "alpha_est": float(s["alpha_est"]),
            "bias_rate": float(s["bias_rate"]),
            "p_target": float(s["p_target"]),
            "T": int(s["T"]),
            "m1": int(s["m1"]),
            "m2": int(s["m2"]),
            "n_reps": int(s["n_reps"]),
            "drop_first_for_fit": int(NEW_DROP_FIRST),
            "theoretical_rate": float(s["theoretical_rate"]),

            "observed_rate_mean_W1": float(s["observed_rate_mean_W1"]),
            "observed_rate_median_W1": float(s["observed_rate_median_W1"]),
            "observed_rate_std_W1": float(s["observed_rate_std_W1"]),

            "observed_rate_mean_MMD": float(s["observed_rate_mean_MMD"]),
            "observed_rate_median_MMD": float(s["observed_rate_median_MMD"]),
            "observed_rate_std_MMD": float(s["observed_rate_std_MMD"]),

            "slopes_W1": g["slope_W1"].astype(float).tolist(),
            "slopes_MMD": g["slope_MMD"].astype(float).tolist(),
            "observed_rates_W1": g["observed_rate_W1"].astype(float).tolist(),
            "observed_rates_MMD": g["observed_rate_MMD"].astype(float).tolist(),
        })

    with open(OUT_RESULTS_JSON, "w", encoding="utf-8") as f:
        json.dump(all_results, f, indent=2)

    print("\nRecomputed rates with drop_first =", NEW_DROP_FIRST)
    print(summary_df[
        [
            "alpha_real",
            "bias_rate",
            "p_target",
            "theoretical_rate",
            "observed_rate_mean_W1",
            "observed_rate_median_W1",
            "observed_rate_mean_MMD",
            "observed_rate_median_MMD",
        ]
    ].to_string(index=False))

    print(f"\nSaved rep-level recomputed results to:\n  {OUT_REP_CSV}")
    print(f"Saved summary table to:\n  {OUT_SUMMARY_CSV}")
    print(f"Saved JSON results to:\n  {OUT_RESULTS_JSON}")

    # New: remake only the last repetition's individual rate plots per combo
    remake_last_rep_rate_plots(recomputed_rep_df)


if __name__ == "__main__":
    main()
```

### ECDF

```python
# ECDF

from pathlib import Path
import json
import numpy as np
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

# ============================================================
# Update sizing (Publication-Quality Style)
# ============================================================
import matplotlib as mpl 
mpl.rcParams["font.sans-serif"] = ["Arial"]
mpl.rcParams["font.size"] = 12
mpl.rcParams["axes.titlesize"] = 26
mpl.rcParams["axes.labelsize"] = 26
mpl.rcParams["legend.fontsize"] = 17 # was 15
mpl.rcParams["legend.title_fontsize"] = 16
mpl.rcParams["xtick.labelsize"] = 24
mpl.rcParams["ytick.labelsize"] = 24

# Cleaner editable text in PDF/SVG
mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["svg.fonttype"] = "none"


# ============================================================
# User settings
# ============================================================

RUN_DIR = Path("logs/ECDF/biased_alphaReal_h0=0.5_T=3000_m1=50_nreps=100_20260611-211935")
REP_TO_COLLECT = 99
DROP_FIRST = 200

OUT_DIR = RUN_DIR / f"figures_single_rep_{REP_TO_COLLECT:03d}_pdf_regenerated"

# Which metrics to regenerate
MAKE_W1 = True
MAKE_MMD = True
MAKE_DENSITY = True
MAKE_TV = False


# ============================================================
# Helpers
# ============================================================

def fmt_float(x):
    return f"{float(x):.2f}"


def plot_density_pdf(xg, f_true_g, f_hat_final, p, alpha, q, save_path):
    plt.figure(figsize=(7, 4))
    plt.plot(xg, f_true_g, linewidth=2, label="True density")
    plt.plot(xg, f_hat_final, linewidth=2, label="Estim. density")
    plt.xlabel("x")
    plt.ylabel("density")
    # plt.title(
    #     f"Estimated density: p={p:.2f}, "
    #     rf"$\alpha_{{real}}$={alpha:.2f}, q={q:.2f}"
    # )
    plt.grid(True, axis="y", ls="--", alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.savefig(save_path, bbox_inches="tight")
    plt.close()


def plot_rate_pdf(M, values, metric_name, p, alpha, q, drop_first, save_path):
    M = np.asarray(M, dtype=float)
    values = np.asarray(values, dtype=float)

    valid = np.isfinite(M) & np.isfinite(values) & (M > 0) & (values > 0)

    log_M = np.full_like(M, np.nan, dtype=float)
    log_val = np.full_like(values, np.nan, dtype=float)

    log_M[valid] = np.log10(M[valid])
    log_val[valid] = np.log10(values[valid])

    plt.figure(figsize=(6, 4))
    plt.plot(log_M, log_val, "o-", alpha=0.5, markersize=3, label="Distr. Loss")

    fit_mask = valid.copy()
    fit_mask[:drop_first] = False

    if np.count_nonzero(fit_mask) > 5:
        x_fit = log_M[fit_mask]
        y_fit = log_val[fit_mask]
        slope, intercept = np.polyfit(x_fit, y_fit, 1)
        y_pred = slope * x_fit + intercept
        plt.plot(x_fit, y_pred, "r--", linewidth=2, label=f"Fit slope: {slope:.3f}")

    if metric_name == "W1":
        metric_tex = r"W_1"
    elif metric_name == "MMD":
        metric_tex = r"\mathrm{MMD}"
    elif metric_name == "TV":
        metric_tex = r"\mathrm{TV}"
    else:
        metric_tex = metric_name

    plt.xlabel(r"$\log_{10}$(M$_n$)")
    plt.ylabel(rf"$\log_{{10}}({metric_tex})$")
    # plt.title(
    #     rf"${metric_tex}$ Scaling: p={p:.2f}, "
    #     rf"$\alpha_{{real}}$={alpha:.2f}, q={q:.2f}"
    # )

    plt.legend()
    plt.grid(True, which="both", ls="--", alpha=0.3)
    plt.tight_layout()
    plt.savefig(save_path, bbox_inches="tight")
    plt.close()


# ============================================================
# Main
# ============================================================

def main():
    OUT_DIR.mkdir(parents=True, exist_ok=True)

    npz_files = sorted((RUN_DIR / "rep_data").glob(f"*/rep={REP_TO_COLLECT:03d}_data.npz"))

    if not npz_files:
        raise FileNotFoundError(
            f"No rep={REP_TO_COLLECT:03d}_data.npz files found under {RUN_DIR / 'rep_data'}"
        )

    print(f"Found {len(npz_files)} rep data files.")

    for npz_path in npz_files:
        data = np.load(npz_path, allow_pickle=True)

        metadata = json.loads(str(data["metadata_json"]))

        p = float(metadata["p_target"])
        alpha = float(metadata["alpha_real"])
        q = float(metadata["bias_rate"])

        base = f"p={fmt_float(p)}_alpha={fmt_float(alpha)}_q={fmt_float(q)}"

        M = data["M"]
        xg = data["xg"]
        f_true_g = data["f_true_g"]

        if MAKE_DENSITY:
            if "f_hat_history" in data:
                f_hat_final = data["f_hat_history"][-1]
            elif "f_hat_final" in data:
                f_hat_final = data["f_hat_final"]
            else:
                raise KeyError(f"No f_hat_history or f_hat_final in {npz_path}")

            density_path = OUT_DIR / f"density_saved_{base}.pdf"
            plot_density_pdf(xg, f_true_g, f_hat_final, p, alpha, q, density_path)
            print(f"Saved {density_path}")

        if MAKE_W1:
            W1 = data["W1_for_slope"] if "W1_for_slope" in data else data["W1"]
            w1_path = OUT_DIR / f"{base}_W1.pdf"
            plot_rate_pdf(M, W1, "W1", p, alpha, q, DROP_FIRST, w1_path)
            print(f"Saved {w1_path}")

        if MAKE_MMD:
            MMD = data["MMD_for_slope"] if "MMD_for_slope" in data else data["MMD"]
            mmd_path = OUT_DIR / f"{base}_MMD.pdf"
            plot_rate_pdf(M, MMD, "MMD", p, alpha, q, DROP_FIRST, mmd_path)
            print(f"Saved {mmd_path}")

        if MAKE_TV:
            TV = data["TV_for_slope"] if "TV_for_slope" in data else data["TV"]
            tv_path = OUT_DIR / f"{base}_TV.pdf"
            plot_rate_pdf(M, TV, "TV", p, alpha, q, DROP_FIRST, tv_path)
            print(f"Saved {tv_path}")

    print("\nDone.")
    print(f"PDFs saved to: {OUT_DIR}")


if __name__ == "__main__":
    main()
```
